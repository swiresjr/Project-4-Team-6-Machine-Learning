```python

# Fetch dataset
predict_students_dropout_and_academic_success = fetch_ucirepo(id=697) 

# Extract feature data
X = predict_students_dropout_and_academic_success.data.features 
filtered_data = X[X["Mother's qualification"].isin([1, 37, 19, 38, 3, 34, 2])]

# Map numeric values to education level names
education_mapping = {
    1: "High School Only",
    37: "4th, 5th grade",
    19: "9th, 10th, 11th grade",
    38: "6th, 7th, 8th grade",
    3: "AA or certification",
    2: "Bachelor's"
}

# Check available columns
#print(X.columns)

filtered_data["Mother's qualification"] = filtered_data["Mother's qualification"].map(education_mapping)


if "Mother's qualification" in X.columns:
    # Count occurrences of each education level
    moms_schooling = filtered_data["Mother's qualification"].value_counts()

    # Plot the bar chart
    moms_schooling.plot.bar(color='tab:blue')
    plt.xlabel("Highest Level of Education")
    plt.ylabel("Number of Mothers")
    plt.title("Mother's Education Level")
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.show()

```

```python


# Filter the data to include only these levels
filtered_moms = X[X["Mother's qualification"].isin(education_mapping.keys())]
filtered_dads = X[X["Father's qualification"].isin(education_mapping.keys())]
```

```python
### Group by Mother's education and Student Outcome, count occurrences
grouped_moms = filtered_moms.groupby(["Mother's qualification", "Student Outcome"]).size().unstack(fill_value=0)

# Apply the existing mapping to replace numbers with education types
grouped_moms.index = grouped_moms.index.map(education_mapping) 

# Plot the graph
grouped_moms.plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Mother's Education Level vs Student Outcome")
plt.xlabel("Mother's Education Level")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.legend(title="Student Outcome")
plt.show()
```

```python
# Group by Mother's education and Student Outcome, count occurrences
grouped_dads = filtered_dads.groupby(["Father's qualification", "Student Outcome"]).size().unstack(fill_value=0)

grouped_dads.index = grouped_dads.index.map(education_mapping) 

# Plot the graph
grouped_dads.plot(kind="bar", stacked=True, figsize=(10, 6))
plt.title("Father's Education Level vs Student Outcome")
plt.xlabel("Father's Education Level")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.legend(title="Student Outcome")
plt.show()
```
